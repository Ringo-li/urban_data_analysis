```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

#### 1.加入文字、图例、网格等

```python
x = np.linspace(0,5)

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.plot(x,x ** 2)
ax1.plot(x,x ** 3)

ax1.text(3,15,'$y=x^2$')
ax1.text(3,40,'$y=x^3$')

ax1.grid(True,alpha=0.6)
ax1.legend(['$y=x^3$','$y=x^3$'],loc=2,ncol=1)

plt.title('趋势图',fontsize=15)

plt.show()
```
