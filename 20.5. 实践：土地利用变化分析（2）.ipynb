#### 思路：
>1.在总表里添加需要的新增和减少用地的字段<br>
>2.写一个函数，获取每个村变化面积总和<br>
>3.写一个函数，将第二步获取的面积数更新到总部的目标字段<br>
>4.写执行代码，将增加和减少字段都更新到总表<br>
#### 工具：
>添加字段：AddField_management<br>
>列出要素类：ListFeatureClass<br>
>游标：SearchCursor/UpdataCursor<br>

#### 1.添加字段

```python
import arcpy
```

```python
fc = r'D:\rui\code_analysis\homework\file\GIS\class1904\class1904_basic_data.gdb\XZQ_vill_277'
field_add = ['red00_13','add00_13']

for i in field_add:
    arcpy.AddField_management(fc,i,'DOUBLE')
print('done')
```

#### 2.循环获取各村的面积

```python
def get_vill_area(fc):
    with arcpy.da.SearchCursor(fc,['SHAPE_Area']) as cursor:
        vill_sum = 0
        for row in cursor:
            vill_sum += row[0]
    return vill_sum
```

#### 2.更新值

```python
def updata_dt(fc,use_field,vill,vill_sum):
    with arcpy.da.UpdateCursor(fc,use_field,"XZQMC = \'" + vill + "\'") as cursor:
        for row in cursor:            
            row[0] = vill_sum
            cursor.updateRow(row)
```

#### 3.执行代码

```python
fc = r'D:\rui\code_analysis\homework\file\GIS\class1904\class1904_basic_data.gdb\XZQ_vill_277'
use_field = ['red00_13','add00_13']
for fd in use_field:
    print(r'D:\rui\code_analysis\homework\file\GIS\class1904\{}.gdb'.format(fd.split('00')[0]))
```

```python
fc = r'D:\rui\code_analysis\homework\file\GIS\class1904\class1904_basic_data.gdb\XZQ_vill_277'
use_field = ['red00_13','add00_13']
for fd in use_field:
    gdbs = r'D:\rui\code_analysis\homework\file\GIS\class1904\{}.gdb'.format(fd.split('00')[0])
    arcpy.env.workspace = gdbs
    vill_list = arcpy.ListFeatureClasses()
    for vill in vill_list:
#         clasuse = "XZQMC = \'" + vill + "\'"
        vill_sum = get_vill_area(vill)
        updata_dt(fc,[fd],vill,vill_sum)
        print('{} finished!'.format(vill.encode('utf-8')))
print('#########################')
print('All finished!')
```
