```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json(r'Project\art_crawl\article\article\spiders\author_use.json')
df.drop(axis=1,columns='Unnamed: 0',inplace=True)
df['journal'] = '城市规划'
```

#### 1.各作者发表论文数

```python
author_art = df.pivot_table(index='author',columns='journal',values='participation',aggfunc='sum')
```

```python
author_art.sort_values(by='城市规划',ascending=False).head(3)
```

```python
df[['author','author_num']].sort_values(by='author_num',ascending=False).head(3)
```

```python
df['kw_union'] = df.keywords.apply(lambda x:','.join(x))
```

```python
df[df['abstract'].str.contains('智能')]
```

```python
lst = ['计算机','大数据','智慧','智能','图像识别','算法','机器学习']
col = ['abstract','kw_union','title']
targ = []
for x in col:
    for y in range(len(df)):
        cont = df[x][y]
        for z in lst:
            if z in cont:
                targ.append(df.title[y])
set(targ)           
```

```python
df_ol = df[['title','date']].drop_duplicates(subset='title').reset_index(drop=True)
df_ol
```

```python
targ_year = pd.merge(pd.DataFrame(set(targ),columns=['title']),df_ol)
```

```python
targ_year.date = targ_year.date.apply(lambda x:x.split('年')[0])
```

```python
targ_year.groupby(['date']).count()
```

```python
targ_res = df[df.title.apply(lambda x:x in list(targ_year.title))]
```
