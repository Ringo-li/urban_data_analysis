```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

#### 一、figure

```python
x = [1,2,3,4,5]
y1 = [i ** 2  for i in x]
y2 = [i ** 2 * 2 for i in x]

plt.figure(figsize=(5,5))
plt.plot(x,y1)
plt.plot(x,y2)

plt.show()
```

#### 二、axes（子图大小一致）

#### 1. plt.subplot(nrows,ncols,plot_number)
+ nrows:图中的行数
+ nclos:图中的列数
+ plot_number:图标编号（从1开始）

```python
plt.figure()

plt.subplot(2,3,1)
plt.bar(x,y1)
plt.title('A')

plt.subplot(2,3,2)
plt.title('B')

plt.subplot(2,3,3)
plt.title('C')

plt.subplot(2,3,4)
plt.title('D')

plt.subplot(2,3,6)
plt.plot(x,y2)
plt.title('E')
          
plt.tight_layout()#加这行自动调节
plt.show()
```

```python
plt.figure()

name = list('ABCDEF')
for i in range(len(name)):
    plt.subplot(2,3,i+1)
    plt.title(name[i])

plt.tight_layout()
# plt.show()
```

#### 2. plt.subplots(nrows,nclos,sharex=True,sharey=True)一次性完成网格
+ nrows:图中的行数
+ nclos:图中的列数
+ sharex和sharey，fig和ax作为整体，侧重于否共享标签

```python
# plt.figure()

# plt.subplots(2,3,sharex=True,sharey=True)
# plt.tight_layout()
# plt.show()
fig, ax = plt.subplots(2,3,sharex=True,sharey=True)#fig和ax作为整体,需定义两个值
ax[0,1].scatter(x,y2)
ax[0,1].set_title('scatter')
ax[1,0].bar(x,y1)
ax[1,0].set_title('bar')
```

#### 三、axes（子图大小不一致）

#### 1.plt.subplot()

```python
plt.subplot(2,3,1)
plt.subplot(2,3,2)
plt.subplot(2,3,3)
plt.subplot(2,1,2)
plt.tight_layout()
```

#### 2.add_subplot()

```python
fig = plt.figure()
fig.add_subplot(2,3,1)
fig.add_subplot(2,3,2)
fig.add_subplot(2,3,4)
fig.add_subplot(2,3,5)
fig.add_subplot(1,3,3)
plt.tight_layout()
```

#### 3.plt.subplot2grid(shape,loc,rowspan=1,colspan=1,fig=None)

```python
plt.figure()
plt.subplot2grid((2,3),(0,0),colspan=2)
plt.plot(x,y1)
plt.title('A')

plt.subplot2grid((2,3),(0,2),rowspan=2)
plt.title('B')

plt.subplot2grid((2,3),(1,0))
plt.title('C')

plt.subplot2grid((2,3),(1,1))
plt.title('D')

plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.subplot2grid((3,3),(0,0))
plt.title('A')

plt.subplot2grid((3,3),(0,1),colspan=2)
plt.title('B')

plt.subplot2grid((3,3),(1,0),colspan=2,rowspan=2)
plt.barh(x,y1)
plt.title('C')

plt.subplot2grid((3,3),(1,2))
plt.title('D')

plt.subplot2grid((3,3),(2,2))
plt.title('E')

plt.tight_layout()
plt.show()
```

#### 4.plt.Gridspec(nrows,nclos,hspace=0.5,wspace=0.5)类
+ nrows,nclos行数列数
+ hspace=0.5,wspace=0.5调整间距
+ 对类实例化后，使用索引、切片进行子图创建

```python
grid = plt.GridSpec(2,3,hspace=0.3,wspace=0.3)

plt.subplot(grid[0,0:2])
plt.barh(x,y1)
plt.title('A')

plt.subplot(grid[1,0])
plt.subplot(grid[1,1])
plt.subplot(grid[0:2,2])
plt.show()
```

```python
grid = plt.GridSpec(3,3,hspace=0.3,wspace=0.3)

plt.subplot(grid[0,0])
plt.subplot(grid[0,1:])
plt.subplot(grid[1:,:-1])
plt.subplot(grid[1:,2:])
plt.show()
```
