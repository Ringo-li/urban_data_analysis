```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import arrow
plt.style.use('ggplot')

%matplotlib inline
```

```python
df = pd.read_json('./file/suzhou_land.json')
```

### 1.数据清洗

```python
#备份数据
df2 = df.copy()
```

```python
#将乱码去除
df2['GDFS'] = df2['GDFS'].apply(lambda x:x.replace('\t','').replace('\n','').replace('\r','').strip())
```

```python
#修改列名
df2.rename(columns={'XZQM':'行政区名','XMMC':'项目名称','XMWZ':'项目位置','HTMH':'合同编号','GDFS':'供地方式'}).head(2)
```

```python
#用导入的arrow库对时间格式重新定义
df2['YDJDSJ'] = df2['YDJDSJ'].apply(lambda x:arrow.get(x).format('YYYY'))
df2['QDRQ'] = df2['QDRQ'].apply(lambda x:arrow.get(x).format('YYYY'))
```

#### 1.2筛选2009-2018住宅用地

```python
#选QDRQ中包含2009-2018字符串的项
sz = df2[df2['QDRQ'].isin(['2009','2010','2011','2012','2013',
                          '2014','2015','2016','2017','2018'])]
```

```python
#选中TDYT中居住用地的项
sz = sz[sz['TDYT']=='其他普通商品住房用地']
```

### 2.数据分析

#### 2.1供地数据分析

```python
dm = sz.groupby(by='QDRQ').count()['CJJG']
dm.plot(kind='bar',alpha=0.7,figsize=(8,6))
plt.title('\n2009-1018居住用地供地宗数',fontsize=18)
plt.xlabel('年份',fontsize=16)
plt.ylabel('宗数（单位：总）',fontsize=16)
plt.xticks(rotation=0)
for i in range(10):
    plt.text(i-0.13,dm[i]+3,dm[i])
```

```python
dm = sz.groupby(by='QDRQ').sum()['MJ']
dm.plot(kind='bar',alpha=0.7,figsize=(8,6))
plt.title('\n2009-1018居住用地供地面积',fontsize=18)
plt.xlabel('年份',fontsize=16)
plt.ylabel('面积（单位：公顷）',fontsize=16)
plt.xticks(rotation=0)
for i in range(10):
    plt.text(i-0.15,dm[i]+20,'{:.0f}'.format(dm[i]))
```

```python
dm = sz.groupby(by='QDRQ').sum()['CJJG']
dm.plot(kind='bar',alpha=0.7,figsize=(8,6))
plt.title('\n2009-1018居住用地出让金额',fontsize=18)
plt.xlabel('年份',fontsize=16)
plt.ylabel('出让金额',fontsize=16)
plt.xticks(rotation=0)
for i in range(10):
    plt.text(i-0.4,dm[i]+200*1000,'{:.0f}亿元'.format(dm[i]/10000))
```

#### 2.2供应方式分析

```python
#对单独一列统计
dk = sz['GDFS'].groupby(by=sz['GDFS']).count()
```

```python
dk.plot(kind='pie',figsize=(6,6),fontsize=14,autopct='%.2f%%')
plt.ylabel('')
```
