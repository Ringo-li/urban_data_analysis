#### 1.geopandas安装
>geopandas有很多依赖包，使用下面命令进行安装<br>
>conda install --channel conda-forge geopandas<br>
+ 在运行过程中遇到如下报错：<br>
>ImportError: The descartes package is required for plotting polygons in geopandas.<br>
+ 解决方法：<br>
>conda install descartes

#### 3.geopandas基本操作

```python
import geopandas as gpd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
gpd.__version__#查看版本
```

```python
regions = gpd.GeoDataFrame.from_file('./file/shp/jiangsu_city.shp',encoding='gbk')
type(regions)
```

```python
regions.geometry.head()#显示区域坐标
```

```python
regions.plot()#图像可视化
plt.show()
```

```python
#调整图像大小，设置边线颜色，设置图像填充颜色,线条宽度
regions.plot(figsize=(6,6),edgecolor='k',facecolor='none',linewidth=0.5)
plt.grid(alpha=0.4)
# plt.xticks([])#不显示x轴标签，网格也会消失
# plt.yticks([])#不显示y轴标签，网格也会消失
plt.show()
```
