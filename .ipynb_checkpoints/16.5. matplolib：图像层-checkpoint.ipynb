```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### 1.颜色

#### 参数：
>color或者c<br>
#### 颜色值设置方式：
+ 1.别名：
    - color = 'b' #b = blue
+ 2.合法的HTML颜色名：；
    - color = 'red'
+ 3.HTML16进制字符串
    - color = '#eeefff'
+ 4.归化到[0,1]的GRB元组 X/255
    - color = (0.3,0.4,0.5)

```python
x = [1,2,3,4,5]
y1 = [i * 2 for i in x]
y2 = [i * 3 for i in x]
y3 = [i * 4 for i in x]
y4 = [i * 5 for i in x]

plt.figure()
plt.plot(x,y1,c='r')
plt.plot(x,y2,c='blue')
plt.plot(x,y3,c='#ee2ff2')
plt.plot(x,y4,c=(0.3,0.4,0.5))

plt.show()
```

```python
rg = range(1,4)
values = [2,5,9]

plt.figure()
plt.bar(rg,values,color=['r',(0.2,0.7,0.2),'#eea432']) #bar中不能用c简写color
plt.show()
```

### 2.线型

#### 2.1线型样式
##### 参数：
>linestyle或者ls
##### 线型值：
>点式线：':' 虚线：'--' 点划线'-.' 实线：'-' 空白线：''

```python
x = np.arange(0,7)

plt.figure()
plt.plot(x,x + 1,c='r',ls='-')
plt.plot(x,x + 2,c='g',ls='--')
plt.plot(x,x + 3,c='b',ls='-.')
plt.plot(x,x + 4,c='pink',ls=':')
plt.plot(x,x + 5,c='r',ls='')

plt.show()
```

#### 2.2线宽
##### 参数：
>linewidth或者lw

```python
x = np.arange(0,7)

plt.figure()
plt.plot(x,x + 1,c='r',ls='-',lw=1)
plt.plot(x,x + 2,c='g',ls='--',lw=2)
plt.plot(x,x + 3,c='b',ls='-.',lw=4)
plt.plot(x,x + 4,c='pink',ls=':',lw=8)
plt.plot(x,x + 5,c='r',ls='',lw=1)

plt.show()
```

#### 2.3线型标记
##### 参数：
>marker #marksize可以调节大小
##### 值：
>正方形:'s' 五边形:'p'  六边形:'H''h' 八边形:'8' 圆形:'o' 菱形:'D''d' 数字1-4

```python
x = np.arange(0,7)

plt.figure()
plt.plot(x,x + 1,c='r',ls='-',lw=1,marker='s',markersize=8)
plt.plot(x,x + 2,c='g',ls='--',lw=2,marker='4',markersize=12)
plt.plot(x,x + 3,c='b',ls='-.',lw=4,marker='h',markersize=12)
plt.plot(x,x + 4,c='pink',ls=':',lw=8,marker='8',markersize=15)
plt.plot(x,x + 5,c='r',ls='-',lw=1,marker='o',markersize=8)

plt.show()
```

#### 2.4复合参数：颜色-样式-标记

```python
x = np.arange(0,7)

plt.figure()
plt.plot(x,x + 1,'g:2')
plt.plot(x,x + 2,'r--s')
plt.plot(x,x + 3,'r-.h')

plt.show()
```
