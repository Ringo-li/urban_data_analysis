#### 1.基本设置

```python
plt.figure(figsize=(5,5))
plt.scatter(xx,xx + 0.25 * pls)
plt.plot(xx,xx ** 2)

#####################################
#标题
plt.title('散点图叠加曲线图',fontsize=20)
#####################################
#xy标签
plt.xlabel('$x-label$',size=15)
plt.ylabel('y-label',size=20)
#####################################
#xy刻度
# plt.xticks([-1,0,1],size=15)
# plt.yticks([-1,0,1],size=15)
#####################################
#xy刻度2
plt.xticks([-1,0,1],('Tom', 'Dick', 'Harry'),size=15)
plt.yticks([-1,0,1],['a','b','c'],size=15)
#####################################

plt.show()
```

#### 2.多字图设置

```python
x = np.linspace(0,5,100)

fig = plt.figure(figsize=(8,4))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.plot(x,x ** 2)
ax2.plot(x,x ** 3)

ax1.set_xlabel('x-label',size=15)
ax1.set_ylabel('y-label',size=15)
ax1.set_xticks([i for i in range(-1,7)])
ax1.set_xticklabels(['v' + str(i) for i in range(-1,7)],size=12)
ax1.set_title('图1')

ax2.set_xlabel('x-label',size=15)
ax2.set_ylabel('y-label',size=15)
ax2.set_title('图2',size=15)


plt.tight_layout()
plt.show()
```

#### 3.共享坐标轴twinx(),twiny()

```python
x = np.linspace(0,5,100)

fig = plt.figure(figsize=(4,4))
ax1 = fig.add_subplot(111)
ax1.plot(x,x ** 2,'r')
ax1.set_ylabel('$y=x^2$',size=12)
ax2 = ax1.twinx()#twinx的作用是创建一个ax2，和ax1共用同一个x轴
ax2.plot(x,x ** 3)
ax2.set_ylabel('$y=x^3$',size=12)
ax2.set_title('平方和立方对比图',size=15)


plt.show()
```

#### 4.显示其他几个象限

##### Step1：
+ ax.spines[].set_color()#[]中依次为top/bottom/left/right
+ ax.spines[].set_lineweight()

##### Step2：
+ ax.spines[].set_position(['data',0])#设置极轴位置，选择data类型，在0位置
+ ax.spines[].set_position(['axes',0.5])#设置极轴位置，选择axes类型，在0位置


```python
fig, ax = plt.subplots()

#设置不需要的轴为无色
ax.spines['right'].set_color('none')
ax.spines['top'].set_color(None)
ax.spines['bottom'].set_color('y')
ax.spines['left'].set_color('g')
#设置需要的轴起点位置
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

xx = np.linspace(-1.,1.,100)
ax.plot(xx,xx ** 2,'r')
ax.set_title('$y=x2$')
ax.plot(xx,xx ** 3,'b')

plt.show()
```

#### 6.刻度显示范围

>plt.ylim()<br>
>plt.xlim()

```python
city = ['chengdu','shanghai','hangzhou','suzhou']
gdp = [1200,2100,1700,1500]
x_lb = [1,2,3,4]

plt.figure()
plt.bar(range(1,5),gdp,alpha=0.7,width=0.5)
plt.xticks(x_lb,city)
plt.ylim(0,2500)

plt.xlabel('城市',fontsize=15)
plt.ylabel('GDP（百亿）',fontsize=15)
plt.title('信息产业前四城GDP对比',fontsize=20)

plt.show()
```

#### 7.图像中的文字显示

>plt.text(x,y,conts)

```python
city = ['chengdu','shanghai','hangzhou','suzhou']
gdp = [1200,2100,1700,1500]
x_lb = [1,2,3,4]

plt.figure()
plt.bar(range(1,5),gdp,alpha=0.7,width=0.5)
plt.xticks(x_lb,city)
plt.ylim(0,2500)

plt.xlabel('城市',fontsize=15)
plt.ylabel('GDP（百亿）',fontsize=15)
plt.title('信息产业前四城GDP对比',fontsize=20)

for i in range(4):
    plt.text(x_lb[i]-0.1,gdp[i]+100,gdp[i])
    
plt.show()
```

#### 8.图例和网格

+ 图例：plt.legend([name1,name2],loc=1,ncol=2)<br>
>#图例名称列表<br>
>#loc图例显示位置<br>
>#ncol图例以几列显示<br>
+ 网格：plt.grid(True)

```python
x = np.linspace(0,10,50)#默认50
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,
        c='r',
        linestyle='--',linewidth=1,
        marker='4',markersize=10
        )
plt.plot(x,y2,
        c='b',
        linestyle='-',linewidth=2,
        marker='3',markersize=10
        )

plt.legend(['$y=sin(x)$','$y=cos(x)$'],loc=1,ncol=1)
plt.grid(True,alpha=0.6)
plt.show()
```
