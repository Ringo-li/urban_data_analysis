```python
import pickle
import os


os.chdir('c:/Users/Administrator/Desktop/')
f = open('定西牛肉面馆.txt','r',encoding='utf8')
print(os.getcwd())
f.seek(0)
info = []
n = 0
for line in f.readlines():
    st1 = line.split(sep=':')
    st2 = st1[1].split(sep=',')
    name = st1[0]
    lnt = st2[0]
    lat = st2[1]
    address = st2[2].strip()
    adict = {'name':name,'lnt':lnt,'lat':lat,'address':address}
    info.append(adict)
    n += 1
    
pick = open('data.pkl','wb')
pickle.dump(info,pick)
pick.close()
print(info)

print('\n A total of %d converted'%n)
f.close()
```

```python
import string
import os
alist = list(range(1,11))
print(alist)
blist = list(string.ascii_lowercase[:10])
print(blist)


os.chdir('c:/Users/Administrator/Desktop/')
os.getcwd()
f = open('write_test.txt','w',encoding='utf8')
adict = dict(zip(alist,blist))
st = list(adict.items())
print(st)
print(len(st))
for i in range(len(st)):
    line = str(st[i][0]),',',st[i][1],'\n'
    print(line) 
    f.writelines(line)
f.close()
```
