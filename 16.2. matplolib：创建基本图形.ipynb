```python
import matplotlib.pyplot as plt
import numpy as  np

#plt.rcParams['font.sans-serif']=['SimHei']
#matplotlib中文无法显示，自定义字体
%matplotlib inline
#魔法工具，使图形在notebook中显示
```

#### 1.折线图

#### matplotlib函数：plot(x,y)

```python
x = list(range(1,6))
y1 = [i * 2 for i in x]
y2 = [m ** 2 for m in x]
```

```python
plt.figure(figsize=(5,5))

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

#### 2.柱状图

#### matplotlib函数 bar(left,height)

```python
dt = np.arange(1,4)
values = [2,4,9]

plt.figure(figsize=(5,5))

plt.bar(dt,values)
plt.show()
```

#### 2.1复合型bar图,width参数

```python
dt = list(range(1,4))
values1 = [1,3,5]
values2 = np.random.randint(3,8,3)

plt.figure(figsize=(6,6))

plt.bar(dt,values1,width=0.4)
plt.bar(dt,values2,bottom=values1,width=0.4)
plt.show()
```

#### 3.横向柱状图

#### matplotlib 函数 barh(bottom,width)

```python
categories = list('ABCDE')
values = np.random.randint(2,9,5)

plt.figure(figsize=(5,5))
plt.barh(categories,values,height=0.6)
plt.show()
```

```python
categories = list('ABCDE')
values = np.random.randint(2,9,5)

plt.figure(figsize=(6,6))
plt.barh(np.arange(len(categories)),values,height=0.6)
plt.yticks(np.arange(len(categories)),['类别%s' % i for i in categories])
plt.show()
```

#### 4.直立方

#### matplotlib函数 hist(x)

```python
rng = np.random.RandomState(seed=0)
column_data = rng.normal(42,3,1000)#创建1000个 正态（高斯）分布随机数，平均值42，标准差3

plt.figure()
plt.hist(column_data)
plt.show()
```

#### 5.饼图

#### matplotlib函数 pie(sizes)

```python
counts = [95,5]

plt.figure()
plt.pie(counts)
plt.show()
```

#### 5.1基本参数设置

```python
counts = [95,5]

plt.figure()
plt.pie(counts,
        colors = ['green','pink'],
        labels = ['CityA','CityB'],
        startangle = 90,#旋转角度，默认从0度开始
        autopct = '%.2f%%'
       )

plt.legend()#图例
plt.show()
```

#### 6.散点图

#### matplotlib 函数 scatter(x,y)

```python
x = np.arange(20)
y = np.arange(50,70) + (np.random.random(20) * 10)

plt.figure()
plt.scatter(x,y)
plt.show()
```

#### 6.1散点大小变化

```python
city = ['chengdu','xian','suzhou','wuhan','hangzhou']
gdp = [16000,9000,17000,15000,21000]
size = np.array(gdp)/10

color = ['r','b','g','black','orange']

plt.figure()
plt.scatter(range(1,6),gdp,c=color,s=size,alpha=0.5)#s气泡大小，alpha透明度
plt.xticks(range(1,6),city)#x轴标签
plt.grid(True)#是否显示网格线
plt.show()
```

#### 7.填充

```python
x = np.linspace(0,1)#从0-1等分50份
y = np.sin(4 *  np.pi * x) * np.exp(-5 * x)

plt.fill(x,y)
plt.grid(True,linewidth = 0.5,alpha = 0.5)
plt.show()
```
