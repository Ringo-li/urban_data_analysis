```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

#### 1.Series对象图像可视化.plot()方法

```python
fig = plt.figure(figsize=(14,4))
#创建100个符合高斯分布的数组
dt_x = np.random.randn(100)

ax1 = fig.add_subplot(1,3,1)

#创建一个series对象，索引data_range设置起始时间和周期
ts = pd.Series(dt_x,index=pd.date_range('2020-1-1',periods=100))

#创建关于Series对象的图形，默认为线型图
ts.plot()

ax2 = fig.add_subplot(1,3,2)
ts2 = ts.cumsum()#cumsum--pd对象累计和
ts2.plot(kind='line')#创建关于Series对象的图形，line为默认值

ax3 = fig.add_subplot(1,3,3)
#ts.plot(kind='hist')#plot中的kind参数选择哪种图，或者可以用plot.hist()
ts.plot.hist()

plt.tight_layout()
plt.show()
```

#### 2.DataFrame对象图像可视化.plot()方法

```python
dt_y = np.random.randn(100,4)#创建一个2维数组,100行，4列
df = pd.DataFrame(dt_y,index=ts.index,columns=list('ABCD'))#创建一个DataFrame对象
df = df.cumsum()#DataFrame对象各列累计求和
```

```python
df.plot()#创建DataFrame对象的默认图形
plt.show()
```

```python
df.plot(subplots=True,figsize=(3,6))#打开子图subplots，设置图像大小figsize
plt.show()
```

```python
#设置图像布局方式layout，打开网格，关闭x轴标签,关闭图例，每个子图加图名,共享y轴
df.plot(subplots=True,figsize=(8,5),layout=(2,2),use_index=False,grid=True,
       title=['A图','B图','C图','D图'],
       legend=False,
       sharey=True)
plt.show()
```
